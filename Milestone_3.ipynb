```python
import os
import numpy as np
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from nltk.tokenize import word_tokenize

# Load your pre-trained Word2Vec model
model = keyedvectors.load_word2vec_format('GoogleNews-vectors-negative300.bin/GoogleNews-vectors-negative300.bin' , binary=True, limit=100000)

# Input and output folders
input_folder = 'videos_scripts'
output_folder = 'Embeddings/'

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Iterate through all .srt files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".srt"):
        input_file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.npy")

        # Read .srt file
        with open(input_file_path, "r", encoding="utf-8") as file:
            subtitles = file.read()

        # Tokenize the subtitles into words
        tokens = word_tokenize(subtitles)

        # Generate embeddings for the entire text
        text_embedding = np.array([model[word] for word in tokens if word in model.key_to_index])
        print(text_embedding , "\n")

        # Save the embeddings to the output file in binary format
        np.save(output_file_path, text_embedding)

print("Embeddings generated and saved successfully.")
```
